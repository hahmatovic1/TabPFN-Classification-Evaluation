```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 5))
sns.countplot(x='G3', data=df)
plt.title('Distribucija Konačnih Ocjena (G3)')
plt.xlabel('Ocjena (G3)')
plt.ylabel('Broj učenika')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Atributi koje želimo prikazati (isključivo numerički za barplot)
attributes = ['absences', 'G1', 'G2', 'G3']

plt.figure(figsize=(12, 6))

for i, attr in enumerate(attributes, 1):
    plt.subplot(2, 2, i)
    sns.barplot(x='studytime', y=attr, data=df, palette='Set1')
    plt.title(f'Prosječna vrijednost: {attr} u odnosu na studytime')
    plt.xlabel('Vrijeme učenja (studytime)')
    plt.ylabel(attr)

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Atributi koje želimo vizualizirati
attributes = ['G1', 'G2', 'G3']

plt.figure(figsize=(12, 6))
sns.set(style="whitegrid", font_scale=1.1)

for i, attr in enumerate(attributes, 1):
    plt.subplot(1, 3, i)
    sns.histplot(data=df, x=attr, hue='studytime', bins=15,
                 element='step', stat='count', common_norm=False,
                 palette='Set2')
    plt.title(f'Distribucija: {attr}')
    plt.xlabel(attr)
    plt.ylabel('Broj učenika')

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Pretpostavljamo da je df tvoj student performance DataFrame
# Pravimo novu kolonu G3_kategorija
def kategorija(ocjena):
    if ocjena <= 8:
        return 'nizak'
    elif ocjena <= 14:
        return 'srednji'
    else:
        return 'visok'

df['G3_kategorija'] = df['G3'].apply(kategorija)

# Atributi koje ćemo prikazati
attributes = ['G1', 'G2', 'studytime', 'failures', 'absences']

# Stil i figura
sns.set(style="white", font_scale=1.2)
plt.figure(figsize=(15, 10))

# Paleta boja
custom_palette = {
    'nizak': 'lightcoral',
    'srednji': 'gold',
    'visok': 'lightgreen'
}

for i, attr in enumerate(attributes, 1):
    plt.subplot(2, 3, i)
    sns.histplot(
        data=df, x=attr, hue='G3_kategorija',
        bins=30, stat='count', common_norm=False,
        palette=custom_palette, alpha=0.6
    )
    plt.title(f'Distribucija atributa: {attr}', fontsize=14, weight='bold')
    plt.xlabel(attr)
    plt.ylabel('Broj učenika')
    plt.grid(False)

plt.tight_layout()
plt.legend(title='G3 nivo', labels=['Nizak', 'Srednji', 'Visok'], bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

fig, axes = plt.subplots(1, 2, figsize=(18, 5))

# Prvi graf - famsup
sns.countplot(x='famsup', data=df, palette='Set3', ax=axes[0])
axes[0].set_title('Podrška porodice pri učenju')
axes[0].set_xlabel('Porodična podrška (famsup)')
axes[0].set_ylabel('Broj učenika')
axes[0].set_xticklabels(['Ne', 'Da'])

# Drugi graf - health
sns.countplot(x='health', data=df, palette='Set2', ax=axes[1])
axes[1].set_title('Samoprocjena Zdravlja')
axes[1].set_xlabel('Ocjena zdravlja (1 = loše, 5 = odlično)')
axes[1].set_ylabel('Broj učenika')

plt.tight_layout()
plt.show()
```
