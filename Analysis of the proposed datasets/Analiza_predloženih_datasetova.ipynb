# Breast Cancer dataset

```python
#Uƒçitavanje datoteke
import pandas as pd
df = pd.read_csv("data_Breast_Cancer.csv")
#Ukalananje poslednje kolone
df = df.drop(columns=["Unnamed: 32"])
df
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(6, 4))
sns.countplot(x='diagnosis', data=df)
plt.title('Distribucija Dijagnoza (Maligno vs Benigno)')
plt.xlabel('Dijagnoza')
plt.ylabel('Broj uzoraka')
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Atributi koji se prikazuju
attributes = [
    'radius_mean',
    'perimeter_mean',
    'area_mean',
    'concavity_mean',
    'concave points_mean'
]

# Postavke dizajna
sns.set(style="white", font_scale=1.2)
plt.figure(figsize=(15, 10))

# Prilagoƒëena paleta boja
custom_palette = {"B": "skyblue", "M": "lightcoral"}


for i, attr in enumerate(attributes, 1):
    plt.subplot(2, 3, i)
    sns.histplot(
        data=df, x=attr, hue='diagnosis',
        bins=30, stat='count', common_norm=False,
        palette=custom_palette, alpha=0.6
    )
    plt.title(f'Distribucija atributa: {attr}', fontsize=14, weight='bold')
    plt.xlabel(attr)
    plt.ylabel('Broj uzoraka')
    plt.grid(False)

plt.tight_layout()
plt.legend(title='Dijagnoza', labels=['Benigno', 'Maligno'], bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

# Prima Indian Diabetes Dataset

```python
#Uƒçitavanje datoteke
import pandas as pd
df = pd.read_csv("diabetes.csv")
df
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(6, 4))
ax = sns.countplot(x='Outcome', data=df)
plt.title('Distribucija Dijagnoza (Dijabetes vs Bez dijabetesa)')
plt.xlabel('Dijagnoza')
plt.ylabel('Broj uzoraka')
plt.xticks([0, 1], ['Bez dijabetesa', 'Sa dijabetesom'])
# Dodavanje procenta iznad stupaca
total = len(df)
for p in ax.patches:
    count = int(p.get_height())
    percentage = 100 * count / total
    ax.annotate(f'{count} ({percentage:.1f}%)',
                (p.get_x() + p.get_width() / 2, p.get_height()),
                ha='center', va='bottom', fontsize=10)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Atributi koje ≈æelimo prikazati
attributes = [
    'Glucose',
    'BloodPressure',
    'BMI',
    'Insulin',
    'Age'
]

# Postavke dizajna
sns.set(style="white", font_scale=1.2)
plt.figure(figsize=(15, 10))

# Prilagoƒëena paleta boja
custom_palette = {0: "skyblue", 1: "salmon"}

for i, attr in enumerate(attributes, 1):
    plt.subplot(2, 3, i)
    sns.histplot(
        data=df, x=attr, hue='Outcome',
        bins=30, stat='count', common_norm=False,
        palette=custom_palette, alpha=0.6
    )
    plt.title(f'Distribucija atributa: {attr}', fontsize=14, weight='bold')
    plt.xlabel(attr)
    plt.ylabel('Broj uzoraka')
    plt.grid(False)

plt.tight_layout()
plt.legend(title='Dijagnoza', labels=['Bez dijabetesa', 'Sa dijabetesom'], bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

```python
import numpy as np

# Kolone koje sadr≈æe implicitne nedostajuƒáe vrijednosti
cols_with_missing = ['Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI']

# Zamjena nula sa NaN
df[cols_with_missing] = df[cols_with_missing].replace(0, np.nan)

# Popunjavanje NaN vrijednosti medianom, bez inplace
for col in cols_with_missing:
    df[col] = df[col].fillna(df[col].median())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Atributi koje ≈æelimo prikazati
attributes = [
    'Glucose',
    'BloodPressure',
    'BMI',
    'Insulin',
    'Age'
]

# Postavke dizajna
sns.set(style="white", font_scale=1.2)
plt.figure(figsize=(15, 10))

# Prilagoƒëena paleta boja
custom_palette = {0: "skyblue", 1: "salmon"}

for i, attr in enumerate(attributes, 1):
    plt.subplot(2, 3, i)
    sns.histplot(
        data=df, x=attr, hue='Outcome',
        bins=30, stat='count', common_norm=False,
        palette=custom_palette, alpha=0.6
    )
    plt.title(f'Distribucija atributa: {attr}', fontsize=14, weight='bold')
    plt.xlabel(attr)
    plt.ylabel('Broj uzoraka')
    plt.grid(False)

plt.tight_layout()
plt.legend(title='Dijagnoza', labels=['Bez dijabetesa', 'Sa dijabetesom'], bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

# Wine Quality dataset

```python
#Uƒçitavanje datoteke
import pandas as pd
df = pd.read_csv("winequalityN.csv")
df
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(8, 5))
ax = sns.countplot(x='quality', hue='quality', data=df, palette='Set3', legend=False)
plt.title('Distribucija klasa ‚Äì Ocjena kvaliteta vina')
plt.xlabel('Ocjena kvaliteta (quality)')
plt.ylabel('Broj uzoraka')

# Dodavanje procenta iznad svake kolone
total = len(df)
for p in ax.patches:
    count = int(p.get_height())
    percentage = 100 * count / total
    ax.annotate(f'{count} ({percentage:.1f}%)',
                (p.get_x() + p.get_width() / 2, p.get_height()),
                ha='center', va='bottom', fontsize=9)

plt.tight_layout()
plt.show()

```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Atributi koje ≈æelimo prikazati (prilagodi po ≈æelji)
attributes = [
    'alcohol',
    'volatile acidity',
    'sulphates',
    'citric acid',
    'density'
]


# Postavke dizajna
sns.set(style="white", font_scale=1.2)
plt.figure(figsize=(16, 10))

# Boje za vi≈°eklasa klasifikaciju
palette = sns.color_palette("tab10", n_colors=df['quality'].nunique())

for i, attr in enumerate(attributes, 1):
    plt.subplot(2, 3, i)
    sns.histplot(
        data=df, x=attr, hue='quality',
        bins=30, stat='count', common_norm=False,
        palette=palette, alpha=0.7,
        multiple='stack'
    )
    plt.title(f'Distribucija atributa: {attr}', fontsize=14, weight='bold')
    plt.xlabel(attr)
    plt.ylabel('Broj uzoraka')
    plt.grid(False)

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Grupisanje kvaliteta u tri klase
def map_quality(q):
    if q in [3, 4]:
        return 'Nizak kvalitet'
    elif q in [5,6]:
        return 'Srednji kvalitet'
    else:
        return 'Visok kvalitet'

# Nova kolona
df['quality_grouped'] = df['quality'].apply(map_quality)

# 2. Prikaz broja uzoraka po novim klasama
class_counts = df['quality_grouped'].value_counts().sort_index()
total = len(df)

print("Distribucija po grupama:")
for k, v in class_counts.items():
    print(f"{k}: {v} uzoraka ({100 * v / total:.1f}%)")

# 3. Grafiƒçki prikaz raspodjele
plt.figure(figsize=(7, 5))
ax = sns.countplot(x='quality_grouped', data=df, palette='Set2')
plt.title('Distribucija kvaliteta vina (3 klase)')
plt.xlabel('Kategorija kvaliteta')
plt.ylabel('Broj uzoraka')

# Dodavanje brojeva i procenata iznad stubaca
for p in ax.patches:
    count = int(p.get_height())
    percentage = 100 * count / total
    ax.annotate(f'{count} ({percentage:.1f}%)',
                (p.get_x() + p.get_width() / 2, p.get_height()),
                ha='center', va='bottom', fontsize=10)

plt.tight_layout()
plt.show()
```
